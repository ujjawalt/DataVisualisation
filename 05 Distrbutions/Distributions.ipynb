# Distributions

## Seting up Environment

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading Dataset

```python
iris_data = pd.read_csv("Iris.csv")
```

Check the dataset loaded properly by printing the first five rows.

```python
iris_data.head()
```

## Histograms
Histograms are one of the fundamental way to represent a distribution:

```python
sns.distplot(a=iris_data['PetalLengthCm'], kde=False)
```

The above is the distribution plot. Next is the Histplot of the same data column:

```python
sns.histplot(data=iris_data["PetalLengthCm"], kde=False)
```

The difference in **distplot** and **histplot** being:
- distplot are distribution plot whereas histplot are data FREQUENCY plot.
- distplot are used to plot the data in continues manner whereas histplot represent discrete values.

## Density plot
Outlines the distribution of data with a continues curve.

```python
sns.kdeplot(data=iris_data["PetalLengthCm"], shade=True)
```

## 2D KDE plots
We can plot more than one columns by creating a _two dimensional (2D) KDE plot_.

```python
sns.jointplot(x=iris_data['PetalLengthCm'], y=iris_data['SepalWidthCm'], kind='kde', shade=True)
```

In the plot above the color coding represent the density, that is how likely we are to see different combinations of _Sepal width_ and _Petal lenght_, where darker parts of the figure are more likely.

Also note that in additon to the KDE plot in the center, 
> the curve at the top of the figure is a KDE plot for _Petal Lenght_ plotted on x-axis, and

> the curve on the right of the figure is a KDE plot for _Sepal Width_ plotted on y-axis.

## Color coded plots
We can plot the three classes of the Iris flower. For this we need to separate the data into three different dataframes. 

```python
iris_set = pd.DataFrame(data=iris_data.iloc[:50,:])
iris_vers = pd.DataFrame(data=iris_data.iloc[50:101,:])
iris_virg = pd.DataFrame(data=iris_data.iloc[101:,:])
```

We will create three different _histogram_ for each species by using the _sns.histplot_. We can differentiate the data using different labels.

```python
plt.figure(figsize=(10,6))
sns.distplot(a=iris_set['PetalLengthCm'], label='Iris-Setosa', kde=False)
sns.distplot(a=iris_vers['PetalLengthCm'], label='Iris-Versicolor', kde=False)
sns.distplot(a=iris_virg['PetalLengthCm'], label='Iris-Virginica', kde=False)

plt.title("Iris species Histogram based on Petal Lenghts")

plt.legend()
```

The legend does not automatically appear in this plot, to force show it (for any plot type), we can add _plt.legend()_.

## KDE plot for each species based on Petal Length:

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=iris_set['PetalLengthCm'], label='Iris-Setosa', shade=True)
sns.kdeplot(data=iris_vers['PetalLengthCm'], label='Iris-Versicolor', shade=True)
sns.kdeplot(data=iris_virg['PetalLengthCm'], label='Iris-Virginica', shade=True)

plt.title("Iris species KDE Plot based on Petal Lenghts")

plt.legend()
```

We can observe that there is clearly some pattern based on the "Petal Lenght" in the data. 
- Iris-setosa is a different class itself.
- Iris-versicolor & Iris-virginica have almost same Petal lenght. 
- If the Petal lenght of any Iris flower is less than 2 Cm it can be calssified as Iris-setosa right away.
