# Plot types and Custom Styles

## Choosing Type of Plot

It is really important to use the correct plot to get the real useful information from it. Based on the different types of data plot can differ:

* **Trends:** The changes in the data over time are defined as _Trends_.
> **Lineplot** (sns.lineplot) are best to represent the trends  oveer a period oof time. (Note: Multiple lines can be used to show different data on same plot

* **Relationship:** To show the _Relationship_ between the observations following plots can be used:
> **Bar Charts** (sns.barplot) can be used to compare different variables.

    > **Heatmaps** (sns.heatmap) is great tool to show the correlation with some added colors in a tabular form of numbers.
    
    > **Scatter plot** (sns.scatterplot) can show relationship between two continues variables. In addition we can add a third variable if we use color coding.
    
    > **Regression Line** (sns.regplot) is great way to visualise the linear relationship (if any present) between two variables. <br>
    _sns.lmplot_ can be used to show multiple regression lines.
    
    > **Categorical scatter plot** (sns.swarmplot) can be useful when we've to show relationship between a continues vairable and a categorical variable.
    
* **Distribution:** Distribution shows the trend that a data is following. It can useful when we want to show what value to expect in the obervation, also the probability of occurrence.
> **Histograms** (sns.histplot) used to show the distribution of single observation.

    > **KDE Plots:**
        1. sns.kdeplot can be used to plot a smooth KDE plot of one or more variable.
        2. sns.jointplot can display a 2D KDE plot along with the corresponding KDE plot of each 
           variable.
    

# Custom Styles using Seaborn
Seaborn have five different styles:
- darkgrid
- whitegrid
- dark
- white
- ticks

### Importing Libraries

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Loadig Data

```python
data = pd.read_csv("spotify.csv")
data.head()
```

### Line plot data (default style: white)

```python
plt.figure(figsize=(16,6))

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data)
```

### Changing Style with Seaborn : darkgrid

```python
plt.figure(figsize=(16,6))

sns.set_style('darkgrid')

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data)
```

### Changing Style with Seaborn : whitegrid

```python
plt.figure(figsize=(16,6))

sns.set_style('whitegrid')

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data)
```

### Changing Style with Seaborn : dark

```python
plt.figure(figsize=(16,6))

sns.set_style('dark')

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data)
```

### Changing Style with Seaborn : ticks

```python
plt.figure(figsize=(16,6))

sns.set_style('ticks')

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data)
```
