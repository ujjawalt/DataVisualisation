# Line Chart

## Importing Libraries

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Impoting Dataset
[Dataset](https://www.kaggle.com/pranaykankariya/spotify) spotify.csv focus on five popular songs from 2017 to 2018 tracked global daily streamss on the music streaming service Spotify.
- "Shape of You", by Ed Sheeran
- "Despacito", by Luis Fonzi
- "Something Just Like This", by The Chainsmokers and Coldplay
- "HUMBLE.", by Kendrick Lamar
- "Unforgettable", by French Montana

```python
data_spotify = pd.read_csv("spotify.csv", index_col="Date", parse_dates=True)
```

## Examine Data
Take a look at first few rows of data.

```python
data_spotify.head()
```

- NaN short for "Not a Number" appears in place of missing entries.

```python
data_spotify.tail()
```

## PLot the Data as Line Chart
Now that data is loaded and we've took a look at it, its time to plot the data:

```python
sns.lineplot(data=data_spotify)
```

Let's add some more components to above plot :

```python
# Set  wdth & Height of figure.
plt.figure(figsize=(16,6))

# Add Title
plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data=data_spotify)
```

## Plot a Subset of the Data
So far we've plot a line for all the column in the data. Alternatively we can plot a subset of data, we'll see how:

We begin by taking a look at name of all the columns in the data:

```python
list(data_spotify.columns)
```

Let's say we wish to plot initial two columns of the dataset:

```python
plt.figure(figsize=(16,6))

plt.title("Daily Global Streams of Popular Songs in (2017-2018)")

sns.lineplot(data = data_spotify['Shape of You'])

sns.lineplot(data = data_spotify['Despacito'])

# Add horizintal axis label
plt.xlabel("Date")
plt.ylabel(None)
```
