# Line Chart 

## Importing Libraries

```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Loading Dataset

```python
museum_data = pd.read_csv("Museum_Visitors.csv", index_col="Month", parse_dates=True)
museum_data.head()
```

## List of all the musuems in the dataset.

```python
list(museum_data.columns)
```

## Dropping Museums we're not concerned with:

```python
museum_data = museum_data.drop(['Gateway to Nature Center', 'Hellman Quon', 'IAMLA','Pico House ', 
                               'Visitor Center/ El Tranquilo Gallery','Museum of Social Justice', 
                               'Biscailuz Gallery/ PK Outdoor Exhibit'], axis=1)
```

```python
museum_data.head()
```

```python
plt.figure(figsize=(16,6))

plt.title("Daily Museam visit")

sns.lineplot(data=museum_data)
```

## Assess Seasonality
Let's take Avila Adobe and plot a line chart to see the monthly visitor trend.

```python
plt.figure(figsize=(16,6))

plt.title("Monthly Visitors to Avila Adobe")

sns.lineplot(data=museum_data['Avila Adobe'])

plt.xlabel("Date")
```

## Conclusion:
- The Avila Adobe museum have a drop in visitor around the month of September-Febuary.
- Around March-July the musuem have high traffic.
